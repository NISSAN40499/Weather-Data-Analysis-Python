# Question two: find the number of time when the weather is exactly "clear"

```python
clear_count=dataset[dataset["Weather"]=="Clear"]["Weather"].count()
print("The number of time when the weather is clear is: ", clear_count)
```

# Question three: find the number of times when the wind speed was exactly 4?

```python
windspeed_count = dataset[dataset["Wind Speed_km/h"]==4]["Wind Speed_km/h"].count()
print("Number of time windspeed was exactly 4 is: " , windspeed_count)
```

# Question thirteen: what is the min against each weather condition?
